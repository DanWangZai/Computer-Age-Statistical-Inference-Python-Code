### Raw Data

The raw data is in a text file. We read this raw data. 

In the book Table 8.5, the variable "redshift" and "magnitude" are shown as category. We show the actual value here. The result is the same either we are using the raw value or the categorical value. We have the
use the "use_categorical" to control whether we will use the raw value or the categorical value in the following. 

```python
def get_rid_of_line_change(word):
    return float(''.join(re.compile('[0-9+.]').findall(word)))
```

```python
filename = r'.\data\galaxy.txt'
use_categorical = False
# use_categorical is True, the redshift and magniude are used as categorical value as 1, 2, 3, 4
```

```python
file1 = open(filename,"r+") 

# header
header = file1.readline()
redshift = header.split(' ')
redshift[-1] = get_rid_of_line_change(redshift[-1])
redshift = [float(x) for x in redshift]

ind = 0;
magnitude = []
line = file1.readline();
n_col = len(redshift)
n_row = 0; 
while(len(line)):
    line = line.split(' ')
    line[-1] = get_rid_of_line_change(line[-1])
    # add the magnitude
    magnitude.append(float(line[0]))
    # add the count to total count
    line =  [int(x) for x in line[1:]]
    count = line if n_row == 0 else np.concatenate([count, line])
    n_row += 1
    # read the next line
    line = file1.readline();
file1.close()
data = pd.DataFrame(count.reshape((n_row, n_col)), index = magnitude, columns = redshift)

if use_categorical is True:
    data.index = np.arange(18, 0, -1)
    data.columns = np.arange(1, 16)
data.index.name = 'magnitude'
data.columns.name = 'redshift'
data
```

For the redshift, the larger the value, the farther. (from left to right get farther)

For the magnitude, the large the value, the dimmer. (from bottom to top get dimmer)

### the Poisson Regression 

#### Poisson Regression Model

$Y_i$ is the observed data, it follows Poisson distribution with parameter $\mu_i$, so $Y_i \sim Poisson(\mu_i)$: 

$$P(Y_i = k) = \frac{\exp^{-\mu_i}\mu_i^{k}}{k!}$$

where $\log(\mu_i)$ is assumed to follow a linear model such that $\log(\mu_i) = X_i'\alpha$, where $X_i$ is the explanatory variables, $\alpha$ is the coefficient that need to be estimated.

According to the book, $Y_i$ is the count in each bin, and $X$ is 
$$[r, m, r^2, m^2, r* m]$$

where $r$ is the redshift, and $m$ is the magnitude. 

#### Create X and Y 

```python
df = pd.DataFrame(data.unstack()).rename(columns = {0:'count'}).reset_index()
df['red_square'] = df['redshift']**2
df['mag_square'] = df['magnitude']**2
df['red_mag'] = df['redshift'] * df['magnitude']

df.head()
```

```python
X = df[['redshift', 'magnitude', 'red_square', 'red_mag', 'mag_square']].values
Y = df['count'].values
```

#### Run Poisson Regression

```python
model = PoissonRegressor(alpha = 0,  max_iter=1000, fit_intercept = True)
model.fit(X, Y)
Y0 = model.predict(X)
df['fit'] = Y0
```

#### Deviance

Finally, we calculate the sum of the deviance. 

Let $y$ be the actual count, $\hat{y}$ is the model predicted count. The deviance is:
    $$D(y,\hat{y})= 2 * y\big((\frac{\hat{y}}{y}-1) - \log(\frac{\hat{y}}{y})\big)$$
    
We calculate the sum of deviance as $S=\sum |D_{i}|$.

```python
D = 2* df['count']* (df['fit']/df['count'] - 1 - log(df['fit']/df['count']))
S = np.sum(abs(D))
print("The S is {:d}".format(int(S)))
```
